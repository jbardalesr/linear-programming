```python
from scipy.optimize import linprog
import sympy as sp
from itertools import combinations
import math
import math
import sympy as sp
from IPython.display import display_latex, Latex
```

```python
# problem a
c = [20, 10]

A = [[-30, -17],
     [-5, -9]]

b = [-60, -70]

x0_bounds = (0, None)
x1_bounds = (0, None)

# minimize a standard form
res = linprog(c, A_ub=A, b_ub=b, bounds=[
              x0_bounds, x1_bounds], method="revised simplex")
res
```

```python
def pretty_print(string: str):
    return display_latex(Latex(string))
```

```python
def factible_vertex(A: sp.Matrix, b: sp.Matrix, c: sp.Matrix):
    m, n = A.shape
    comb = tuple(combinations(range(m), n))
    print("Number of combinations:", len(comb))
    # for each combiantion for indexes
    for indexes in comb:
        A_inv = sp.Matrix.inv(A.row(indexes))
        x = A_inv@b.row(indexes)
        show = True
        # for each constraint
        for j in range(m):
            if not A.row(j).dot(x) <= b[j]:
                show = False
                break
        if show: 
            pretty_print(f"$A^{{-1}}={sp.latex(A_inv)}\quad x{indexes}={sp.latex(x)}\quad c={c.dot(x).evalf()}$")
```

```python
## A.row(j).dot(x) <= b[j]: OK 
A = sp.Matrix([[1, 1, 1],
               [-1, 0, 0],
               [0, -1, 0],
               [0, 0, -1]])
b = sp.Matrix([1, 0, 0, 0])

c = sp.Matrix([2, -3, -4])

factible_vertex(A, b, c)
```

```python
A = sp.Matrix([[1, 1],
              [15, -2],
              [5, 1],
              [2, 1],
              [-1, 0],
              [0, -1]])

b = sp.Matrix([4, 0, 5, 3, 0, 0])

c = sp.Matrix([11, 2])

factible_vertex(A, b, c)
```

```python
def basic_solutions(A, b):
    m, n = A.shape
    print("Number of combinations: %d" % math.comb(n, m))
    for comb in combinations(range(n), m):
        try:
            B = A.col(comb)
            B_inv = B.inv()
            x_b = B_inv @ b  # x_n are zeros
            x_labels = sp.Matrix([f"x_{i+1}" for i in comb])
            pretty_print(f"$$B={sp.latex(B)}, \quad \mathbf x_B={sp.latex(x_labels)} = B^{{-1}}\mathbf b= {sp.latex(B_inv)}{sp.latex(b)}={sp.latex(x_b)}$$")
        except Exception as ex:
            print("Column index =", comb, ex)
            continue
```

```python
A = sp.Matrix([[1, 2, 2, 1, 0],
              [2, 1, 2, 0, 1]])

b = sp.Matrix([4, 3])

basic_solutions(A, b)
```
